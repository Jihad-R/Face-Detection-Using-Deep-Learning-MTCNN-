# Importing Libraries

```python
from matplotlib import pyplot 
from mtcnn.mtcnn import MTCNN
```

```python
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle 
```

# Defining that plots the dots and boxes

```python
# Plot a box with the detected Object
def draw_image_with_boxes(file_name, faces):
    data = pyplot.imread(file_name)
    pyplot.imshow(data)
    # get the context to plot boxes
    ax = pyplot.gca()
    
    for face in faces:
        x, y, width, height = face['box']
        rect = Rectangle((x, y), width, height, fill=False, color='yellow')
        ax.add_patch(rect)
        pyplot.axis('off')
        
        for key, value in face['keypoints'].items():
            dot = Circle(value, radius = 4, color='yellow')
            ax.add_patch(dot)
        
    pyplot.show()
```

**Getting the images**

```python
file_name = 'test 5.jpg'
image = pyplot.imread(file_name)
```

**Creating the detector, using default weights**

```python
detector = MTCNN()
```

**Detecting faces in the Image (Pixels)**

```python
faces = detector.detect_faces(image)
```

```python
draw_image_with_boxes(file_name,faces)
```
